```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random
from sklearn import metrics
```

```python
dataset = pd.read_csv("data.csv")
```

```python
dataset
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
```

```python
dataset["EducationSector"]= label_encoder.fit_transform(dataset["EducationSector"])
dataset["IndividualProject"]= label_encoder.fit_transform(dataset["IndividualProject"])
dataset["Gender"]= label_encoder.fit_transform(dataset["Gender"])
dataset["City"]= label_encoder.fit_transform(dataset["City"])
dataset["Influenced"]= label_encoder.fit_transform(dataset["Influenced"])
dataset["MentalDisorder"]= label_encoder.fit_transform(dataset["MentalDisorder"])
dataset["KeyTraits"]= label_encoder.fit_transform(dataset["KeyTraits"])
dataset["ReasonsForLack"]= label_encoder.fit_transform(dataset["ReasonsForLack"])
```

```python
dataset
```

```python
print("Data shape:",dataset.shape)
```

```python
dataset.isnull().sum()
print(dataset.isnull().sum())
```

```python
X = dataset.iloc[:, 0:16].values
x = dataset.iloc[:, 0:16]
y = dataset.iloc[:, 16].values
y_name = dataset.iloc[:, 16:17]
print("X shape:",X.shape)
print("y shape:",y_name.shape)
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
sn.set(font_scale=1)
sn.set_style("darkgrid")
fig_dims = (10, 6)
fig, ax = plt.subplots(figsize=fig_dims)
sn.heatmap(dataset.corr(),annot=True, ax=ax)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(algorithm='auto', leaf_size=30, metric='minkowski',
                     metric_params=None, n_jobs=None, n_neighbors=5, p=2,
                     weights='uniform')
```

```python
classifier_knn.fit(X_train, y_train)
```

```python
y_pred_knn = classifier_knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm_knn = confusion_matrix(y_test, y_pred_knn)
print(cm_knn)
```

```python
sn.heatmap(cm_knn, annot= True)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
ac  = accuracy_score(y_test, y_pred_knn)
pre = precision_score(y_test, y_pred_knn)
re  = recall_score(y_test, y_pred_knn)
f1  = f1_score(y_test, y_pred_knn)
```

```python
print("Summary of KNN")
print("Accuracy    =", ac)
print("Precison    =",pre)
print("Recall      =", re)
print("f1 score    =", f1)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
prob_all_knn = classifier_knn.predict_proba(X_test)
```

```python
probs_knn = prob_all_knn[:, 1]
```

```python
auc_knn = roc_auc_score(y_test, probs_knn)
print('K-NN: ROC AUC=%.3f' % (auc_knn))
```

```python
fpr_knn, tpr_knn, _ = roc_curve(y_test, probs_knn)
```

```python
plt.figure(figsize=(2,1), dpi=300)
sn.set(font_scale=.2)
sn.set_style("darkgrid")
plt.plot(fpr_knn, tpr_knn,linewidth=1,color='darkorange', label = 'AUC = %0.4f' % auc_knn)
plt.plot([0, 1], [0, 1], color='navy', linewidth=1, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("ROC of KNN")
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```
